# Run Simulation

```python
# Run the simulation for several episodes
NUM_EPISODES = 500
sim = ProxySimulation(num_episodes=NUM_EPISODES, output_dir="data")
metrics_history = sim.run()
```

# Analysis: Unbiased Policy Agreement

```python
# Extract metrics
episodes = range(len(metrics_history))
agent0_acc = [m['agent_0']['accuracy'] for m in metrics_history]
agent1_acc = [m['agent_1']['accuracy'] for m in metrics_history]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(episodes, agent0_acc, label='Agent 0 Accuracy')
plt.plot(episodes, agent1_acc, label='Agent 1 Accuracy')
plt.xlabel('Episode')
plt.ylabel('Agreement with Unbiased Policy')
plt.title('Recommender System Bias Over Time (vs Human Proxy)')
plt.legend()
plt.grid(True)
plt.ylim(0, 1.05)
plt.show()
```
